# The Clausius-Clapeyron equation
It represents the saturated vapor pressure as a function of the temperature. Its approximation, valid under typical atmospheric conditions, can be written as

$$ e(T) = c * exp(\frac{aT}{b + T}) $$

where a, b and c are constants and T is the temperature. It tells us how much the saturated vapor pressure in the atmosphere increases for each additional degree of temperature, and therefore how much more vapor it can contain. In numbers it is about 7% more vapor in the atmosphere for each additional degree Celsius. Of course there must be a source of humidity, for example the sea. Italy, being in the middle of the Mediterranean sea, is therefore in an optimal place to have alternating droughts and floods.

```python
import numpy as np
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
```

```python
a = 17.625
b = 243.04
c = 6.1094
T = np.linspace(0, 50, 100) # start = 0, stop = 50, samples = 100
e = c * np.exp(a * T / (b + T))
```

```python
fig = plt.figure(figsize=(10, 5))
ax = fig.add_subplot()
ax.set_title("Saturation Water Pressure (Clausius-Clapeyron)")
ax.set_xlabel("Temperature (°C)")
ax.set_ylabel("Pressure (hPa)");
plt.plot(T, e)
plt.savefig('clausius-clapeyron.png')
```
