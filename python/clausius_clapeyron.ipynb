# Clausius-Clapeyron
L'equazione di Clausius-Clapeyron 

e(T) = c * exp[a / (b + T)]

dove a, b e c sono delle costanti e T e' la temperatura, ci dice di quanto diminuisce la pressione di vapore saturo in atmosfera, per ogni grado in piu' di temperatura, e quindi quanto vapore in piu' puo' contenere. In numeri e' circa il 7% in piu' di vapore in atmosfera per ogni Â°C in piu'. Naturalmente deve esserci una fonte di umidita', per esempio il mare. L'italia trovandosi in mezzo al Mediterraneo e' quindi in una posizione ottimale per avere siccita' e alluvioni alternate.

```python
import numpy as np
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
import seaborn; seaborn.set()
import matplotlib.dates as mdates
```

```python
a = 17.625
b = 243.04
c = 6.1094
T = np.linspace(0, 100, 100) # start = 0, stop = 300, samples = 100
e = c * np.exp(a/(b + T))
```

```python
fig = plt.figure(figsize=(10, 5))
ax = fig.add_subplot()
ax.set_title("Saturation Water Pressure (Clausius-Clapeyron)")
ax.set_xlabel("Temperature (Â°C)")
ax.set_ylabel("Pressure (hPa)");
plt.plot(T, e)
```
