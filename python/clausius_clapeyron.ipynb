# Clausius-Clapeyron
The Clausius-Clapeyron equation 

$$ e(T) = c * exp(\frac{a}{b + T}) $$

where a, b and c are constants and T is the temperature, tells us how much the saturated vapor pressure in the atmosphere decreases, for each additional degree of temperature, and therefore how much more vapor it can contain. In numbers it is about 7% more vapor in the atmosphere for each additional Celsius degree. Of course there must be a source of humidity, for example the sea. Italy, being in the middle of the Mediterranean, is therefore in an optimal place to have alternating droughts and floods.

```python
import numpy as np
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
```

```python
a = 17.625
b = 243.04
c = 6.1094
T = np.linspace(0, 100, 100) # start = 0, stop = 300, samples = 100
e = c * np.exp(a/(b + T))
```

```python
fig = plt.figure(figsize=(10, 5))
ax = fig.add_subplot()
ax.set_title("Saturation Water Pressure (Clausius-Clapeyron)")
ax.set_xlabel("Temperature (°C)")
ax.set_ylabel("Pressure (hPa)");
plt.plot(T, e)
```
