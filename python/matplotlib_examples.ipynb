## Axes
Each subplot has axis that are used to plot the data

```python
import matplotlib.pyplot as plt
import numpy as np
cols = 3
rows = 3
fig, axs = plt.subplots(ncols=cols, nrows=rows, figsize=(5, 5),
                        layout="constrained")
# for each Axes, add an artist, in this case a nice label in the middle...
ax_origin_x = 0.5
ax_origin_y = 0.5
for row in range(rows):
    for col in range(cols):
        axs[row, col].annotate(f'axs[{row}, {col}]', (ax_origin_x, ax_origin_y),
                            transform=axs[row, col].transAxes,
                            ha='center', va='center', fontsize=10,
                            color='darkgrey')
fig.suptitle('plt.subplots()')
```

## Axis origin

```python
fig = plt.figure(figsize=(4, 3), dpi=100)
ax = fig.add_subplot(1, 1, 1)
ax.text(0.5, 0.5, "Luigi", transform=ax.transAxes)
```

## Plot aspect

```python
fig, ax = plt.subplots(figsize=(6,6)) # 600x600 pixels with aspect=1
ax.set_aspect(0.5)
for label in ax.get_xaxis().get_ticklabels():
    label.set_fontweight("bold")
ax.plot(range(10))
```

## Colormaps
A color is represented as an array of three real numbers between 0 and 1. A colormap can be built with an array of arrays of three numbers. The lenght of the array represents the number of colors available in the colormap. Many standard colormaps are available in Matplotlib and can be selected by their name or code. A colormap can be resampled to reduce the number of colors. A custom colormap can be built by defining its color arrays. 

A colormap can be created using hash numbers. A one digit hash number can encode a number of 16 values, and a two digits hash number can encode $16^2=256$ values. So with a string that contains 3 two digits numbers we can represent an RGB color. 

```python
green_color = '#5AAA5F'
light_green_color = '#A7D25C'
yellow_color = '#ECD347'
dark_yellow_color = '#F5BE41'
orange_color = '#F09235'
red_color = '#D93322'
colors = [green_color, light_green_color, yellow_color, dark_yellow_color, orange_color, red_color]
aqi_cmap = ListedColormap('aqi_colormap', colors)
labels = ['Very good', 'Good', 'Medium', 'Poor', 'Very Poor', 'Extremely Poor']
```
